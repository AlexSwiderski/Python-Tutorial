```python
#change bins
plt.hist(sample_data, bins = 30)
```

```python
fig, ax1 = plt.subplots()
ax1.hist(sample_data, bins = 30)

#add a beta dis
a = 5
b = 10
#create object with beta distribution
beta_draws = np.random.beta(a,b)
#set axis labels
ax1.set_ylabel('density')
ax1.set_xlabel('value')
#Add a subplot with beta distribution
ax2 = fig.add_axes([0.125, 0.575, 0.3, 0.3])
ax2.hist(beta_draws)
```

### Plotting time series data
First getting rid of NA values

```python
surveys_na = surveys.dropna()

#grouping by year and species id in new df

yearly_counts = surveys_na.groupby(['year', 'species_id'])['species_id'].count()
print(yearly_counts)
```

```python
(p9.ggplot(data = yearly_counts,
          mapping = p9.aes(x = 'year',
                          y = 'counts',
                          color = 'species_id'))
+ p9.geom_line()
)
```

```python
(p9.ggplot(data = surveys_na, mapping = p9.aes(x = 'weight',
                                              y = 'hindfoot_length',
                                              color = 'species_id'))
+ p9.geom_point(alpha = 0.25)
+ p9.facet_wrap('plot_id')
)
```
